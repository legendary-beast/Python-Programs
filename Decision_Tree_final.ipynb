```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer.keys()
```

```python
cancer['data']
```

```python
cancer['feature_names']
```

```python
cancer['target_names']
```

```python
sns.countplot(x=cancer['target'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = cancer['data']
y = cancer['target']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(criterion='gini',max_depth=5)

dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)
```

```python
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred)) # Shift + Tab
```

```python
single_sample = x_test[0]
dt.predict([single_sample])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
titanic = pd.read_csv('/content/Titanic-Dataset.csv')
titanic.head()
```

```python
titanic.drop(columns=['Cabin','Ticket','Name'])
```

```python
titanic.head()
```

```python
ntitanic = titanic.drop(columns=['PassengerId','Ticket','Fare','Cabin','Name'])
```

```python
ntitanic
```

```python
# # prompt: convert age column in ntitanic dataframe from float to int


count = ntitanic['Age'].isnull().sum().sum()
avg_age = int(ntitanic['Age'].mean())
ntitanic['Age'].fillna(avg_age,inplace = True)
```

```python
ntitanic['Age'] = ntitanic['Age'].astype(int)
count = ntitanic['Age'].isnull().sum().sum()
count
```

```python
# prompt: perform one hot encoding on ntitanic categorical columns

import pandas as pd
ntitanic_encoded = pd.get_dummies(ntitanic, columns=['Sex', 'Embarked'])
print(ntitanic_encoded.head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
le = LabelEncoder()

# Fit the encoder to the categorical features
ntitanic['Sex'] = le.fit_transform(ntitanic['Sex'])
ntitanic['Embarked'] = le.fit_transform(ntitanic['Embarked'])

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(ntitanic.drop('Survived', axis=1), ntitanic['Survived'], test_size=0.25)

# Create a decision tree classifier
clf = DecisionTreeClassifier()

# Train the classifier
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Evaluate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import classification_report,confusion_matrix

cm = confusion_matrix(y_test,y_pred)
report  = classification_report(y_test,y_pred)
print(cm,'/n',report)
```
